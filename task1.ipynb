# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

# Load the dataset

```python
dataset1 = 'DATASETTASK1.csv' 
df = pd.read_csv(dataset1, skiprows=4, delimiter=',')
df
```

# Clean the dataset

```python
df_cleaned = df.drop(columns=['Unnamed: 68']).dropna()
```

# Convert columns from string to numeric where applicable

```python
for col in df_cleaned.columns[4:]:
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
```

# Drop rows where conversion to numeric failed

```python
df_cleaned = df_cleaned.dropna()
```

# Displaying the cleaned data

```python
df_cleaned.head()
```

```python

print(df_cleaned.columns)


print(df_cleaned[['Country Name', '1960', '2000', '2020']].head())
```

# Bar Chart : Top 10 Most Populous Countries in 2020

```python
top_10_countries = df_cleaned[['Country Name', '2020']].sort_values(by='2020', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_countries['Country Name'], y=top_10_countries['2020'], palette='viridis')
plt.title('Top 10 Most Populous Countries in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Histogram : Distribution of Population Growth Rates (2000-2020)

# Calculate the growth rate between 2000 and 2020

```python
df_cleaned['Growth Rate (2000-2020)'] = ((df_cleaned['2020'] - df_cleaned['2000']) / df_cleaned['2000']) * 100
```

# Display the top 20 rows of the dataframe with the growth rate

```python
print(df_cleaned[['Country Name', '2000', '2020', 'Growth Rate (2000-2020)']].head(20))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Growth Rate (2000-2020)'], bins=30, kde=True, color='purple')
plt.title('Distribution of Population Growth Rates (2000-2020)')
plt.xlabel('Growth Rate (%)')
plt.ylabel('Number of Countries')
plt.show()
```

## Histogram - Distribution of Population Growth Rates (2022-2023)


```python

df_cleaned['Growth Rate (2022-2023)'] = ((df_cleaned['2023'] - df_cleaned['2022']) / df_cleaned['2022']) * 100


plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Growth Rate (2022-2023)'], bins=30, kde=True, color='purple')
plt.title('Distribution of Population Growth Rates (2022-2023)')
plt.xlabel('Growth Rate (%)')
plt.ylabel('Number of Countries')
plt.show()
```

## Bar Chart - Top 10 Countries with the Smallest Population in 2023

```python
# Select the top 10 countries with the smallest population in 2023
top_10_smallest_population_2023 = df_cleaned[['Country Name', '2023']].sort_values(by='2023', ascending=True).head(10)

# Plot the bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_smallest_population_2023['Country Name'], y=top_10_smallest_population_2023['2023'], palette='Reds')
plt.title('Top 10 Countries with the Smallest Population in 2023')
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.xticks(rotation=45, ha='right')
plt.show()
```
